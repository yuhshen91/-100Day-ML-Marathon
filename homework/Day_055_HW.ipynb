```python
import numpy as np
import matplotlib
import matplotlib.pyplot as py
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets,cluster
```

```python
iris=datasets.load_iris()
X=iris.data
y=iris.target
```

```python
estimators =[("cluster 8",KMeans(n_clusters=8)),("cluster 3",KMeans(n_clusters=3)),("cluster 2",KMeans(n_clusters=2)),("cluster bad",KMeans(n_clusters=3,init="random"))]
```

```python
fignum=1
titles=["8","3","2","bad"]
for name,est in estimators:
    fig=py.figure(fignum,figsize=(4,3))
    ax=Axes3D(fig,rect=[0,0,.95,1],elev=48,azim=134)
    est.fit(X)
    labels=est.labels_
    ax.scatter(X[:,3],X[:,0],X[:,2],c=labels.astype(np.float),edgecolor="k")
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum-1])
    ax.dict=16
    fignum+=1
fig=py.figure(fignum,figsize=(4,3))
ax=Axes3D(fig,rect=[0,0,.95,1],elev=46,azim=138)
y=np.choose(y,[1,2,0]).astype(np.float)
ax.scatter(X[:,3],X[:,0],X[:,2],c=y,edgecolor="k")
for name,label in [("cls0",0),("cls1",1),("cls2",2)]:
    ax.text3D(X[y==label,3].mean(),X[y==label,2].mean(),X[y==label,2].mean()+2,name,horizontalalignment="center",bbox=dict(alpha=.2,edgecolor="w",facecolor="w"))
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title("Ground")
ax.dist=12
```

```python
iris = datasets.load_iris()
x = iris.data


kmeans_fit = cluster.KMeans(n_clusters = 3).fit(x)


cluster_labels = kmeans_fit.labels_
print("分群結果：")
print(cluster_labels)
print("---")


Y = iris.target
print("真實品種：")
print(Y)
estimators=[("KMeans",cluster_labels),("Real",y)]
fignum=1
titles=["Kmean","Real"]
for name,est in estimators:
    fig=py.figure(fignum,figsize=(4,3))
    ax=Axes3D(fig,rect=[0,0,.95,1],elev=48,azim=134)
    #labels=est.labels_
    ax.scatter(X[:,3],X[:,0],X[:,2],c=est,edgecolor="k")
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum-1])
    ax.dict=16
    fignum+=1
```
