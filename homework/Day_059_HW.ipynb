### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets, decomposition
import numpy as np
import matplotlib.pyplot as py
from mpl_toolkits.mplot3d import Axes3D


digits = datasets.load_digits(n_class=9)
X = digits.data
y = digits.target
np.random.seed(5)
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
pca= decomposition.PCA(n_components=9)
centers=[[1,1],[-1,-1],[1,-1]]
```

```python
pca.fit(X)
X=pca.transform(X)
fig=py.figure(1,figsize=(16,12))
py.clf()
ax=Axes3D(fig,rect=[1,1,.95,1],elev=20,azim=135)
py.cla

#y=np.choose(y,[1,2,0]).astype(np.float)
ax.scatter(X[:,0],X[:,1],X[:,2],c=y,cmap=py.cm.nipy_spectral,edgecolor="k")
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
py.show()
```

```python
digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
np.random.seed(5)
pca= decomposition.PCA(n_components=3)
pca.fit(X)
X=pca.transform(X)
fig=py.figure(1,figsize=(16,12))
py.clf()
ax=Axes3D(fig,rect=[1,1,.95,1],elev=20,azim=135)
py.cla

#y=np.choose(y,[1,2,0]).astype(np.float)
ax.scatter(X[:,0],X[:,1],X[:,2],c=y,cmap=py.cm.nipy_spectral,edgecolor="k",marker="o",s=80)
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
py.show()
```
