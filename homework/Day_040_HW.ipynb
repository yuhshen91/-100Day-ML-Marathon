## 練習時間
試著使用 sklearn datasets 的其他資料集 (boston, ...)，來訓練自己的線性迴歸模型，並加上適當的正則化來觀察訓練情形。

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score
```

```python
boston=datasets.load_boston()
wine=datasets.load_wine()
```

```python
x=boston.data[:,np.newaxis,2]
train_X,test_X,train_y,test_y=train_test_split(x,boston.target,test_size=0.3, random_state=4)
regr=linear_model.LinearRegression()
regr.fit(train_X,train_y)
y_pred=regr.predict(test_X)
print("r2",r2_score(test_y,y_pred))
print("MSE:",mean_squared_error(test_y,y_pred))
plt.scatter(train_X,train_y)
plt.plot(test_X,y_pred)
```

```python
lasso=linear_model.Lasso(alpha=1)
lasso.fit(train_X,train_y)
y_pred=lasso.predict(test_X)
print("r2",r2_score(test_y,y_pred))
print("MSE:",mean_squared_error(test_y,y_pred))
plt.scatter(train_X,train_y)
plt.plot(test_X,y_pred)
```

```python
ridge=linear_model.Ridge(alpha=1)
ridge.fit(train_X,train_y)
y_pred=ridge.predict(test_X)
print("r2",r2_score(test_y,y_pred))
print("MSE:",mean_squared_error(test_y,y_pred))
plt.scatter(train_X,train_y)
plt.plot(test_X,y_pred)
```
